```python
def normalization(dataset_filename, pix_ranges):
    data = scipy.io.loadmat(dataset_filename)
    data_cells = data['dataset']
    amp_imgs = data['amp_imgs']
    depth_imgs = data['depth_imgs']
    f_0 = data['f_0']
    c = data['c']
    dataset = np.empty([5, 70, 172, 224, 9], dtype=complex)
    for i in range(data_cells.shape[0]):
        for j in range(data_cells.shape[1]):
            dataset[i,j,:,:,:]=data_cells[i,j]
    
    # Get shape of dataset
    n_mats = dataset.shape[0]
    n_inst_per_mat= dataset.shape[1]

    # Initialize new dataset
    new_dataset = np.empty((n_mats, n_inst_per_mat), dtype=object)

    for mat_idx in range(n_mats):
        for inst_idx in range(n_inst_per_mat):
            # Retrieve pixel range
            init_row_idx = pix_ranges[2*(inst_idx-1)+1, 2*(mat_idx-1)+1]
            final_row_idx = pix_ranges[2*(inst_idx-1)+1, 2*mat_idx]
            init_col_idx = pix_ranges[2*inst_idx, 2*(mat_idx-1)+1]
            final_col_idx = pix_ranges[2*inst_idx, 2*mat_idx]

            # Obtain reflectivity- and depth-independent features from raw data
            amp_data = amp_imgs[mat_idx, inst_idx]
            amp_norm_dataset = dataset[mat_idx, inst_idx, :, :, 2:] / amp_data

            # Back calculating phase from depth information
            phase_data = 2 * (2 * np.pi) * f_0 * 1e6 * depth_imgs[mat_idx, inst_idx] / c
            amp_size = amp_norm_dataset.shape
            phase_matrix = np.exp(1j * np.repeat(phase_data[:, :, np.newaxis], amp_size[2], axis=2))
            new_dataset[mat_idx, inst_idx] = amp_norm_dataset / phase_matrix

            # Crop Region of Interest from Images
            new_dataset[mat_idx, inst_idx] = new_dataset[mat_idx, inst_idx][init_row_idx:final_row_idx, init_col_idx:final_col_idx, :]

    return new_dataset
```

```python
def frft(x, alpha):
    N = len(x)
    n = np.arange(N)
    F = dft(N)
    phase = -np.pi * alpha * (n ** 2) / N
    D = np.exp(1j * phase)
    X = fft(x)
    Y = np.dot(F, np.multiply(X, D))
    return Y

```

```python
def apply_frft_on_dataset(dataset, alpha):
    n = len(dataset.shape)
    new_shape = list(dataset.shape)
    result = np.zeros(new_shape, dtype=complex)
    
    for i in range(dataset.shape[0]):
        for j in range(dataset.shape[1]):
            for k in range(dataset.shape[2]):
                for l in range(dataset.shape[3]):
                    result[i, j, k, l, :] = frft(dataset[i, j, k, l, :], alpha)
    
    return result

```
